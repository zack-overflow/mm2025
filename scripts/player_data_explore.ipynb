```python
import pickle

# open pickle files
with open('player_data_2024.pkl', 'rb') as f:
    player_data_2024 = pickle.load(f)
with open('player_data_2025.pkl', 'rb') as f:
    player_data_2025 = pickle.load(f)

print("Player data for 2024:")
for player in player_data_2024:
    print(player)

print("\nPlayer data for 2025:")
for i, player in enumerate(player_data_2025):
    print(i, player, player_data_2025[player])
```
